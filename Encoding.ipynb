```python
#without using any kind of encoding
df.replace({'sex':{'Female':1,'Male':0}},inplace= True)
```

```python
df.head()
```

```python
#Using label Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
Encoder = LabelEncoder()
```

```python
df['smoker'] = Encoder.fit_transform(df['smoker'])
```

```python
df['time'] = Encoder.fit_transform(df['time'])
```

```python
df.head()
```

```python
#One-hot Encoding
One_Hot = pd.get_dummies(df['day'],drop_first = True)
```

```python
One_Hot.head()
```

```python
new_df = df.drop('day', axis =1)
```

```python
new_df.head()
```

```python
Encoded = pd.concat([One_Hot,new_df],axis =1)
```
